```python
import numpy as np
import matplotlib.pyplot as plt
```

# Sinusoidal Signal
A sinusoidal signal is a signal that has the form of a sine or cosine wave

* It is periodic that is it repeats itself over time

```python
#Paramters
f = 5 # Frequency of the sinusoid(hz)
fs = 1000 # Sampling frequency (hz)
T = 1 # Duration of the signal
t = np.linspace(0, T, int(fs*T), endpoint = False)
```

(a) Generate a basic sinusoidal signal in the time domain

```python
x = np.sin(2*np.pi * f * t)
plt.figure(figsize=(10,4))
plt.plot(t, x)
plt.title('Time-domain waveform of sinusoidal signal')
plt.xlabel('Time(s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

The sinusoidal signal is generated with a frequency of 5 Hz, sampled at 1000 Hz, for a duration of 1 second

(c) Compute and plot the discrete-time fourier transform(DTFT)

# Discrete- Time fourier transform DTFT
It transforms a discrete-time signal, which is a sequence of numbers into a continuous function of frquency, in simpler terms, it decomposes a digital signal into its constituent frequencies, represented as a continuous spectrum

```python
N=2048 # Number of points of dtft approximation
frequencies = np.fft.fftfreq(N, 1/fs)
X_dtft = np.fft.fft(x, N) #DTFT

plt.figure(figsize=(10,4))
plt.plot(frequencies[:N//2], np.abs(X_dtft[:N//2]) / len(x)) # Plot positive frequencies
plt.title('DTFT (Continuous frequency spectrum)')
plt.xlabel('frequency (HZ)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

(d) Compute plot the discrete fourier tranform DFT

# DFT
DFT is the sampled version of dtft obtained by evaluating dtft at a finite number of frequencies

* while the dtft produces a continuous frequency spectrum, dft produces a discrete frequency spectrum

```python
X_dft = np.fft.fft(x) #DFT using FFT algorithm
frequencies_dft = np.fft.fftfreq(len(x), 1/fs)

plt.figure(figsize=(10,4))
plt.stem(frequencies_dft[:len(x)//2], np.abs(X_dft[:len(x)//2]) / len(x), use_line_collection=True)
plt.title('DFT Discrete Frequecy spectrum')
plt.xlabel('Frequncy(Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

# Exponentially Decaying signal
An exponentially decaying signal is a signal whose amplitude decreases exponentially over time. This means that the signal's value gets smaller and samller as time progresses, eventually approaching zero

```python
tau = 0.1 # Time constant for exponential decay
x = np.exp(-t / tau) # exponentially decaying signal

plt.figure(figsize=(10, 4))
plt.plot(t, x)
plt.title('Time-domain waveform of exponentially decaying signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

(c) Compute and plot the Discrete-Time Fourier Transform (DTFT)

```python
N = 2048
frequencies = np.fft.fftfreq(N, 1/fs)
X_dtft = np.fft.fft(x, N)

plt.figure(figsize=(10, 4))
plt.plot(frequencies[:N//2], np.abs(X_dtft[:N//2]) / len(x))
plt.title('DTFT (Continuous Frequency Spectrum of Exponentially Decaying Signal)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

(d) Compute and plot the Discrete Fourier Transform (DFT)

```python
X_dft = np.fft.fft(x)
frequencies_dft = np.fft.fftfreq(len(x), 1/fs)

plt.figure(figsize=(10, 4))
plt.stem(frequencies_dft[:len(x)//2], np.abs(X_dft[:len(x)//2]) / len(x), use_line_collection=True)
plt.title('DFT (Discrete Frequency Spectrum of Exponentially Decaying Signal)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

(e) Analysis of time-domain vs frequency-domain

* The time-domain waveform shows the amplitude of the signal decreasing exponentially over time.
* In the frequency domain (DTFT and DFT), the signal exhibits a broader spectrum, indicating that the exponential decay introduces multiple frequency components
* The rapid decay of the signal in the time domain results in a wider spread of frequencies in the frequency domain, which is consistent with the uncertainty principle where a sharply localized signal in time leads to a broad frequency spectrum

* Sinusoidal Signals: A pure sine wave has only one frequency. When you look at it in the frequency domain (using DTFT or DFT), you see a sharp spike at that frequency. This means all the energy is focused at that one frequency..

* Composite Signals: When you combine multiple sine waves with different frequencies, the frequency domain shows multiple spikes. Each spike corresponds to one of the sine waves in the mix, making it easy to see which frequencies are present in the signal.

* Exponentially Decaying Signals: For a signal that fades away over time (like an exponentially decaying signal), the frequency domain shows a spread out spectrum. If the signal fades faster, the spread is wider, showing that the signal contains a mix of more frequencies. This is because when a signal changes quickly in time, it affects more frequencies.

* Rectangular Signals: A rectangular pulse (a signal that switches on and off abruptly) creates a frequency pattern that looks like a sinc wave (which has a central peak and ripples on the sides). The sharper the on-off transition in the signal, the more the frequencies are spread out. So, quicker transitions in time lead to more frequencies being involved.
